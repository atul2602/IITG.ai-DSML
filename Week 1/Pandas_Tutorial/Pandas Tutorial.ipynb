### Filtering data

```python
#Multiple conditions
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]   #note '&' sign
```

```python
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
```

```python
new_df  #note that indices are old
```

```python
new_df.to_csv('filtered.csv', index = False)
```

```python
new_df.reset_index(drop = True, inplace = True)   #drop old indices, inplace saves space of assignment
new_df
```

```python
# Filer based on substring
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')]  #negation
```

```python
import re                  #regex

df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]  #flags for F,G caps
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

### Aggregate stats (group_by)

```python
df = pd.read_csv('modified.csv')
```

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending = False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1

df.groupby(['Type 1']).count()['count']
```

```python
df.groupby(['Type 1', 'Type 2']).count()['count']
```

### Working with large amounts of data

```python
for df in pd.read_csv('modified.csv', chunksize = 50):
    print('-----------------------------------------------')
    print(df)
    
```

```python
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('modified.csv', chunksize = 50):
        results = df.groupby(['Type 1']).count()
        
        new_df = pd.concat([new_df, results])

new_df
```
